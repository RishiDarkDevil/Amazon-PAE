## Data Cleaning - Description

I ignore the presence of any HTML structure as of now for ease. Here I clean the description to remove any tags with spaces.

```python
def remove_html_tags(text):
  """Remove html tags from a string"""
  clean = re.compile('<.*?>')
  return ' '.join(re.sub(clean, ' ', text).strip().split())
```

```python
data.DESCRIPTION = data.DESCRIPTION.apply(lambda x: remove_html_tags(x))
```
